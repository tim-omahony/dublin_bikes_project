```python
n = 41
sql = f"""

SELECT number, last_update, available_bikes, available_bike_stands FROM dbbikes_info
where number = {n}
"""

df = pd.read_sql_query(sql, engine)
df['last_update'] = pd.to_datetime(df['last_update'], unit='ms')
df
```

```python
df.set_index('last_update').resample('1d').mean()
```

```python
df.set_index('last_update').resample('1d').mean().plot(y='available_bikes')
```

```python
df.set_index('last_update').resample('1d').mean().plot(y='available_bike_stands')
```

```python
res_df = df.set_index('last_update').resample('1d').mean()
res_df['last_update'] = res_df.index
results = res_df.to_json(orient='records', index=True)
results
```
